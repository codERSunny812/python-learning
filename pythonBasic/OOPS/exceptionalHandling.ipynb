```python
# handling multilple exceptions

try:
    a = 12
    b = 'd'
    c = a+b
except NameError as e1:
    print("Variable is not defined", e1)
except TypeError as e2:
    print("Data type mismatched that's why unable to add",e2)
except Exception as e3:
    print("some other error occured:",e3)
```

```python
try:
    a = int(input("enter a number A :"))
    b = int(input("enter the another number B:"))
    c = a/b
    d = a**b
    e = a+b
except NameError as e1:
    print("Variable is not defined", e1)
except TypeError as e2:
    print("Data type mismatched that's why unable to add",e2)
except ZeroDivisionError as e3:
    print("you cant divide  any number with 0",e3)
except Exception as e4:
    print("some other error occured:",e4)
else:
    print("the value of the about calculation is :\n")
    print(c,d,e)
finally:
    print("the execuution is done") # it will run everytime 

```

#### Custom Exception 


```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("enter your year of birth: "))
age = 2026-year
print(age)

try:
    if age > 20 and  age < 30:
       print("you age is valid for registration")
    else:
       raise dobException
except dobException:
    print("your age is not valid for registration")
finally:
    print("thank you")
```
