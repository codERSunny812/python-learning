```python
class School:
    school_name="ABC High School"

    def __init__(self, student_name, student_age):
        self.student_name = student_name
        self.student_age = student_age
    
    def display_student_info(self):
        print(f"hello everyone, my name is {self.student_name} and i am {self.student_age} and i study in {School.school_name}")

    @classmethod
    def change_school_name(cls, new_name):
        cls.school_name = new_name
```

```python
studnet1 = School("sunny",23)
studnet1.display_student_info()
```

```python
studnet2 = School("vedi",20)
studnet2.display_student_info()
```

```python
School.change_school_name("kendriya vidyalaya")
```

```python
studnet3 = School("pratistha",19)
studnet3.display_student_info()
```
