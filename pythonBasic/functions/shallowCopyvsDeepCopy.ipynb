```python
# = , deep copy and shallow copy in python

list1 = [1, 2, 3, 4, 5]
list2 = list1  # shallow copy
```

```python
# shallow copy 

list1 = [12, 23, 34, 45]
list2 = list1.copy()  # shallow copy using copy() method



print("list1:", list1)
print("list2:", list2)
```

```python
list1[0]=99


print("list1:", list1)
print("list2:", list2)
```

```python
# shallow copy in case of nested list 

lst1 = [[1, 2, 3], [4, 5, 6]]
lst2 = lst1.copy()  # shallow copy using copy() method

print("lst1:", lst1)
print("lst2:", lst2)
```

```python
lst1[1][1]=996

print("lst1:", lst1)
print("lst2:", lst2)
 # you will see that both the lists are affected because only the outer list is copied, the inner lists are still referenced.
```

```python
# deep copy 

import copy 

lst1 = [1,2,3,4,5,6]
lst2 = copy.deepcopy(lst1) # deep copy using copy.deepcopy() method

print("lst1:", lst1)
print("lst2:", lst2)
```

```python
lst1[2]=909

print("lst1:", lst1)
print("lst2:", lst2)    
```

```python
# deep copy in case of nested list

lst1 = [[1, 2, 3], [4, 5, 6]]
lst2 = copy.deepcopy(lst1)  # shallow copy using copy() method

print("lst1:", lst1)
print("lst2:", lst2)
```

```python
lst1[1][0]=78

print("lst1:", lst1)
print("lst2:", lst2)

# you will see that only lst1 is affected, lst2 remains unchanged because deep copy creates a new copy of all nested objects. 
```
