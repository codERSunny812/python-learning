### python decorators

```python
# function call 

def welcome():
    return "welcome to the world"
```

```python
wel = welcome()
del welcome
```

```python
wel
```

```python
# closure 

def parent_func(msg):
    def child_func():
        print("hello from the child function")
        print(msg)
    return child_func
```

```python
parent_func("hi there")()
```

```python
# decorator 

def greet():
   print("good morning")
```

```python
greet()
```

```python
def decorator_func(fun):
    def child_decor_fun():
        print("this is before the function execution")
        fun()
        print("this is after the function execution")
    return child_decor_fun
```

```python
# apply decorator

@decorator_func
def greet():
   print("good morning")
```

```python
greet()
```
