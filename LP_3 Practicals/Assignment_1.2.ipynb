```python
# Load dataset
df = pd.read_csv("./swiggydataset.csv")
print(df)
```

```python
# Take only text column
df = df[['full_text']].dropna()
```

```python
# Function to get sentiment
def get_sentiment(text):
    score = TextBlob(text).sentiment.polarity
    
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Apply sentiment
df['sentiment'] = df['full_text'].apply(get_sentiment)
```

```python
# Show results
df.head()
```

```python
# Count results
print("\nSentiment Count:\n")
print(df['sentiment'].value_counts())
```

```python
def senti_anal(txt):
    score = TextBlob(txt).sentiment.polarity

    if score > 0:
        return "positive sentence"
    elif score < 0:
        return "negative sentence"
    else:
        return "neutral sentence"
```
