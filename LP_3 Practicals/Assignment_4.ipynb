```python
import pandas as pd
import numpy as np
import time

# ================================
# 1. CREATE LARGE DATASET
# ================================
n = 1_000_000  # 1 million rows

data = {
    "id": np.arange(n),
    "value": np.random.randint(1, 100, n),
    "category": np.random.choice(['A', 'B', 'C', 'D'], n)
}

df = pd.DataFrame(data)

# ================================
# 2. IN-MEMORY COMPUTATION
# ================================
start = time.time()

# simple operation: groupby + mean
result_memory = df.groupby("category")["value"].mean()

end = time.time()

memory_time = end - start

print("In-Memory Result:\n", result_memory)
print("In-Memory Time:", memory_time, "seconds")

# ================================
# 3. DISK-BASED COMPUTATION
# ================================
# write to disk
df.to_csv("large_data.csv", index=False)

start = time.time()

# read from disk
df_disk = pd.read_csv("large_data.csv")

# same operation
result_disk = df_disk.groupby("category")["value"].mean()

end = time.time()

disk_time = end - start

print("\nDisk-Based Result:\n", result_disk)
print("Disk-Based Time:", disk_time, "seconds")

# ================================
# 4. COMPARISON
# ================================
print("\n--- COMPARISON ---")
print("In-Memory Time:", memory_time)
print("Disk-Based Time:", disk_time)

if memory_time < disk_time:
    print("In-Memory Computing is Faster")
else:
    print("Disk-Based Computing is Faster")
    
```
