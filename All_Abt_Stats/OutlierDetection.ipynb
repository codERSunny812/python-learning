```python
import numpy as np # importing  numpy  libraray 
```

```python

arr = [2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,38,350,45,199,400]

print("created array is : ", arr)
```

```python
q1 = np.percentile(arr,25)
print("q1 is : ", q1)
q3 = np.percentile(arr,75)
print("q3 is : ", q3)
```

```python
IQR = q3-q1
print("IQR is : ", IQR)
lower_bound = q1 - (1.5 * IQR)
upper_bound = q3 + (1.5 * IQR)

print("lower bound is : ", lower_bound)
print("upper bound is : ", upper_bound)
```

```python
l = []

for i in arr:
    if i > lower_bound and i < upper_bound:
        l.append(i)


arr2 = np.array(l)

print("original array is : ", arr)
print("new array after removing outliers is : ", arr2)
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=arr)
```
