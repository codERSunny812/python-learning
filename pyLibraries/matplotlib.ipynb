# Matplotlib

- It's a plotting library in python. it provide an Object oriented API for embedding plots into the application using general purpose GUI tool kit like Tkinter, Wxpython etc


```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

### Matplot Basic

```python
x = np.linspace(0,10,11)
x
```

```python
y = x**2
y
```

```python
# basic plotting 

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y plot')
```

#### Subplot in matplotlib

```python
# subplotting in python using matplotlib

plt.subplot(2,2,1)

plt.plot(x,y,'r--')
plt.title("Subplot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,2)
plt.scatter(x,y,c='g')
plt.title("Subplot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,3)
plt.bar(x,y,color='b')
plt.title("Subplot 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")



plt.subplot(2,2,4)
plt.plot(x,y,'y^')
plt.title("Subplot 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

#### Object oriented way

```python
fig = plt.figure(figsize=(8,6),dpi=100)

ax1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom,width,height
ax2 = fig.add_axes([0.2,0.4,0.4,0.3])

ax1.plot(x,y)

ax2.plot(x,y)

# we have full control of the position and size of the axes as well as canvas size
```

#### Type of plot

```python
import matplotlib.pyplot as plt

# scatter plot 

plt.scatter(x,y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


```

```python
# histogram 

x = np.random.randint(1,100,30)
plt.hist(x)
```

```python
# box plot 

plt.boxplot(x)
```

#### saving the plot 


```python
fig = plt.figure(figsize=(8,6),dpi=100)

x= np.linspace(0,10,11)
y= x**2

ax1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom,width,height
ax1.plot(x,y)
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.set_title("X vs Y plot")
plt.savefig("basicxyPlot.png") # this is how you save the  file
```
