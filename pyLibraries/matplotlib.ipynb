# Matplotlib

- It's a plotting library in python. it provide an Object oriented API for embedding plots into the application using general purpose GUI tool kit like Tkinter, Wxpython etc


```python
import matplotlib.pyplot as plt 
import numpy as np

# Generate sample data

x= np.arange(0,10)
y=np.arange(1,11)


# scatter plot 

plt.scatter(x,y,c="g")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# plot in python using matplotlib

plt.plot(x,y, color='b', marker='p', linestyle='--')
plt.title("Line Plot Example")
plt.xlabel("X-axis") 
plt.ylabel("Y-axis")
plt.show()
```

```python
# subplotting in python using matplotlib

plt.subplot(2,2,1)

plt.plot(x,y,'r--')
plt.title("Subplot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,2)
plt.scatter(x,y,c='g')
plt.title("Subplot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(2,2,3)
plt.bar(x,y,color='b')
plt.title("Subplot 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")



plt.subplot(2,2,4)
plt.plot(x,y,'y^')
plt.title("Subplot 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# bar plot in python using matplotlib

x = [10,21,31,41]
y=[5,10,15,20]


x2 = [11,31,41,51]
y2=[15,25,35,45]

plt.bar(x,y,color='m')
plt.bar(x2,y2,color='c')
plt.title("Bar Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
