# Categorical Data Distribution Plots using Seaborn

- now we are using the searborn to plot the categorical data 


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# counter plot 

sns.countplot(x=df["sex"],hue=df["smoker"])
```

```python
#bar plot 

sns.barplot(x=df["sex"],y=df["total_bill"],data=df)
```

```python
sns.barplot(x=df["sex"],y=df["total_bill"],data=df,estimator=np.sum)
```

```python
# box plot 

sns.boxplot(x=df["tip"],y=df["day"],data=df,hue=df["sex"])
```

```python
# voilin plot

sns.violinplot(x=df["day"],y=df["total_bill"],data=df)
```

```python
# strip plot 

sns.stripplot(x="day",y="tip",data=df)
```

```python
# swarn plot 

sns.swarmplot(x="day",y="tip",data=df)
```
