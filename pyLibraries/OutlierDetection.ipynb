```python
import numpy as np # importing  numpy  libraray 
```

```python
# creating a custoom array 

arr = np.random.randint(1,100, size=(25))

print("created array is : ", arr)
```

```python
q1 = np.percentile(arr,25)
print("q1 is : ", q1)
q3 = np.percentile(arr,75)
print("q3 is : ", q3)
```

```python
IQR = q3-q1
print("IQR is : ", IQR)
lower_bound = q1 - (1.5 * IQR)
upper_bound = q3 + (1.5 * IQR)

print("lower bound is : ", lower_bound)
print("upper bound is : ", upper_bound)
```

```python
l = []

for i in arr:
    if i < lower_bound or i > upper_bound:
        l.append(i)


arr2 = np.array(l)

print("original array is : ", arr)
print("new array after removing outliers is : ", arr2)
```
