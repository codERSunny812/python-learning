```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IPL.csv')
df

```

```python
# basic info about dataset

df.info()
```

```python
#the no of rows and columns

(row,columns) = df.shape
print(f'The dataset has {row} rows and {columns} columns')
```

```python
#see  how many columns have missing values
df.isnull().sum()
```

```python
# which team has won the most number of matches

match_win = df['match_winner'].value_counts()

sns.barplot(y=match_win.index, x=match_win.values)
```

```python
# Toss decision analysis

sns.countplot(data=df, x='toss_decision')
```

```python
# toss winner vs match winner 

cnt = df[df['toss_winner'] == df['match_winner']].count()
print(f'Toss winner is also match winner {cnt["match_id"]} times')
```

```python
# how do team wins

sns.countplot(data=df, x='won_by')   
```

```python
df.head()
```

```python
# most player of the match awards 

pom = df['player_of_the_match'].value_counts().head(10)
pom
```

```python
sns.barplot(x=pom.values,y=pom.index)
plt.title('Top 10 player of the match awards')
```

```python
# topscorers in the IPL

high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'barh')
```

```python
# most matched played venue

venue_count = df['venue'].value_counts()
venue_count
```

```python
sns.barplot(y = venue_count.index,x = venue_count.values,palette='rainbow')
```

```python
# who won the highest margin of runs

df[df['won_by'] == 'Runs'].sort_values(by = 'margin',ascending=False).head(1)[['match_winner','margin']]

```

```python
# which player had the highest individual score in a match

df[df['highscore'] == df['highscore'].max()][['top_scorer','highscore']]
```
