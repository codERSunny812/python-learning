```python
import pandas as pd
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import minmax_scale
from sklearn.tree import DecisionTreeClassifier
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.drop(columns=['who','adult_male','deck','embark_town','alive','alone','class'],inplace=True)
```

```python
df
```

```python
# train test split 

X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['survived']),df['survived'],test_size=0.25,random_state=42)
```

```python
print("training data is:\n",X_train.sample(3))
print("testing data is :\n",X_test.sample(3))
```

```python
print("training data set of y:\n",y_train.sample(3))
print("testing data set of y:\n",y_train.sample(3))
```

```python
df.isnull().sum()
```

```python
# applying simple imputation 

si_age = SimpleImputer()
si_embark = SimpleImputer(strategy='most_frequent')



X_train_age = si_age.fit_transform(X_train[['age']])
X_train_embark = si_embark.fit_transform(X_train[['embarked']])




X_test_age = si_age.transform(X_test[['age']])
X_test_embark = si_embark.transform(X_test[['embarked']])
```

```python
X_train_embark
```

```python
# one hot encoding Sex and Embarked

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_embark
```

```python
X_train_res = X_train.drop(columns=['sex','age','embarked'])
X_test_res = X_test.drop(columns=['sex','age','embarked'])
```

```python
print(X_train_res.shape)
print(X_train_age.shape)
print(X_train_embarked.shape)
print(X_train_sex.shape)



print(X_test_res.shape)
print(X_test_age.shape)
print(X_test_embarked.shape)
print(X_test_sex.shape)


```

```python

X_train_transformed = np.concatenate((X_train_res, X_train_sex, X_train_age, X_train_embark),axis=1)


X_test_transformed = np.concatenate((X_test_res,X_test_age,X_test_sex,X_test_embark),axis=1)
```

```python

X_test_transformed
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```
